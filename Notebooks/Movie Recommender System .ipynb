```python
# import necessary libraries
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import numpy as np
import pandas as pd
```

```python
# load data files
credits_df = pd.read_csv('../Data/tmdb_5000_credits.csv')
movies_df = pd.read_csv('../Data/tmdb_5000_movies.csv')
```

```python
credits_df.head()
```

```python
movies_df.head()
```

```python
# Merge the data frames
merged_df = credits_df.merge(movies_df, how='inner',on='title')
```

```python
merged_df.head()
```

```python
# Get necessary details for recommender system
movies = merged_df[['movie_id','title','overview','cast','crew','genres','keywords']]
movies.head()
```

```python
def get_cast_list(obj):
    return [cast_dict['name'] for cast_dict in eval(obj)]
        
```

```python
movies['cast'] = movies['cast'].apply(get_cast_list)
```

```python
def get_crew_director(obj):
    return [crew_dict['name'] for crew_dict in eval(obj) if crew_dict['job']=='Director']
```

```python
movies['crew'] = movies['crew'].apply(get_crew_director)
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(get_cast_list)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(get_cast_list)
```

```python
movies.head()
```

```python
# Covert it in list so that we can easly append it with other lists
movies.dropna(inplace=True)
movies['overview'] = movies['overview'].apply(lambda x : x.split(' '))
```

```python
movies.head()
```

```python
# Remove the empty spaces so that it will consider all as unique
# For Example
# Sam Worthington should be like SamWorthington
movies['cast'] = movies['cast'].apply(lambda x:[item.replace(' ','')for item in x])
movies['crew'] = movies['crew'].apply(lambda x:[item.replace(' ','')for item in x])
movies['genres'] = movies['genres'].apply(lambda x:[item.replace(' ','')for item in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[item.replace(' ','')for item in x])
```

```python
movies.head()
```

```python
# Create a new column Tag that contain the long string
movies['tags'] = movies['overview']+movies['cast']+movies['crew']+movies['genres']+movies['keywords']
```

```python
movies.head()
```

```python
fmovies = movies[['movie_id','title','tags']]
```

```python
fmovies.head()
```

```python
fmovies['tags'] = fmovies['tags'].apply(lambda x: ' '.join(x))
```

```python
fmovies.sample(19)
```

# Start the Coding of Recommander system

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
fmovies.shape
```

```python
vector = cv.fit_transform(fmovies['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
fmovies[fmovies['title'] == 'The Lego Movie'].index[0]
```

```python
distances = sorted(list(enumerate(similarity[744])),reverse=True,key = lambda x: x[1])
```

```python
 fmovies.iloc[distances[0][0]].title
```

```python
# Write a function to check the top five recommended movies
def recommended_movies(movie):
    try:
        index = fmovies[fmovies['title'] == movie].index[0]
        distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
        for i in distances[1:6]:
            print(fmovies.iloc[i[0]].title)
    except:
        print("Movie not available")
```

# Testing

```python
recommended_movies("Pirates of the Caribbean: At World's End")
```
